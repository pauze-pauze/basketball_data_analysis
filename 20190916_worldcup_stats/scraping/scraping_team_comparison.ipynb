```python
#各ゲームのteam_comparisonのURL取得
url_list_raw = soup.select("div.participants > a") #gameごとのURL取得
base_url = "http://www.fiba.basketball" 
end_url = "#tab=team_comparison"
url_list = list() #gameのteam_comparisonのURLを格納


for i in range(0, len(url_list_raw)):
    url_join = url_list_raw[i].attrs["href"]
    print(url_join)
    url_link = urljoin(base_url ,url_join) + end_url
    print(url_link)
    url_list.append(url_link)

#.attrs["href"]
```

```python
###team_comparisonのデータ取得
#格納用リスト作成
Name_TeamA = []
Name_TeamB = []
Timing = []
FB_TeamA = []
FB_TeamB = []
PitP_TeamA = []
PitP_TeamB = []


for i in range(0,len(url_list)+1):
    driver.get(url_list[i])
    time.sleep(3)
    r_stats = driver.page_source
    soup = bs4.BeautifulSoup(r_stats, "lxml")
    

    #チーム名
    name_teamA = soup.select("#gamepage_header_game_post > div > div.header-scores_mobile > div.header-scores_details > div.team-A > span")
    print(name_teamA)
    Name_TeamA.append(name_teamA[0].text)
    
    name_teamB = soup.select("#gamepage_header_game_post > div > div.header-scores_mobile > div.header-scores_details > div.team-B > span")
    print(name_teamB)
    Name_TeamB.append(name_teamB[0].text)
    
    #試合タイミング
    timing = soup.select("#gamepage_header_game_post > div > div.header-scores_desktop > div.game-info > div.phase-info > span.phase")
    print(timing)
    Timing.append(timing[0].text)

    #FastBreak
    FB_teamA = soup.select("#gamepage_teamcomparison > div > div > section.key-stats > ul > li.comparison > div.fast-break.compare-container > span.team-A")
    print(FB_teamA)
    FB_TeamA.append(FB_teamA[0].text)    
    
    FB_teamB = soup.select("#gamepage_teamcomparison > div > div > section.key-stats > ul > li.comparison > div.fast-break.compare-container > span.team-B")
    print(FB_teamB)
    FB_TeamB.append(FB_teamB[0].text)
    
    
    #Points in the Paint
    PitP_teamA = soup.select("#gamepage_teamcomparison > div > div > section.key-stats > ul > li.comparison > div.points.compare-container > span.team-A")
    print(PitP_teamA)
    PitP_TeamA.append(PitP_teamA[0].text)
    
    PitP_teamB = soup.select("#gamepage_teamcomparison > div > div > section.key-stats > ul > li.comparison > div.points.compare-container > span.team-B")
    print(PitP_teamB)
    PitP_TeamB.append(PitP_teamB[0].text)
    
   
```

```python
print(len(Name_TeamA))
print(len(Name_TeamB))
print(len(Timing))
print(len(FB_TeamA))
print(len(FB_TeamB))
print(len(PitP_TeamA))
print(len(PitP_TeamB))

```

```python
df_team_comparison = pd.DataFrame({
    "Name_TeamA" : Name_TeamA,
    "Name_TeamB" : Name_TeamB,
    "Timing" : Timing,
    "FB_TeamA" : FB_TeamA,
    "FB_TeamB" : FB_TeamB,
    "PitP_TeamA" : PitP_TeamA,
    "PitP_TeamB" : PitP_TeamB
})
df_team_comparison
```
