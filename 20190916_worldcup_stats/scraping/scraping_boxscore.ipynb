```python
import bs4,requests, time, lxml.html
import pandas as pd
from selenium import webdriver
import re
from urllib.parse import urljoin
```

```python
#seleniumからwebページ読み込み
#options = webdriver.ChromeOptions()

#options.add_argument('--user-agent=Mozilla/5.0 (iPhone; CPU iPhone OS 10_2 like Mac OS X) AppleWebKit/602.3.12 (KHTML, like Gecko) Version/10.0 Mobile/14C92 Safari/602.1')

#driver = webdriver.Chrome(chrome_options=options)

driver = webdriver.Chrome()

driver.get("http://www.fiba.basketball/basketballworldcup/2019/games")

r = driver.page_source

soup = bs4.BeautifulSoup(r, "lxml")
```

```python
#各ゲームのteam_comparisonのURL取得
url_list_raw = soup.select("div.participants > a") #gameごとのURL取得
base_url = "http://www.fiba.basketball" 
end_url = "#tab=boxscore"
url_list = list() #gameのteam_comparisonのURLを格納


for i in range(0, len(url_list_raw)):
    url_join = url_list_raw[i].attrs["href"]
    print(url_join)
    url_link = urljoin(base_url ,url_join) + end_url
    print(url_link)
    url_list.append(url_link)
```

```python
for i in range(0,len(url_list)):
    print(url_list[i])
len(url_list)
```

```python
#格納用リスト作成
Timing = []
Name_TeamA =[]
Min_TeamA = []
Pts_TeamA = []
Fg_TeamA = []
F2g_TeamA = []
F3g_TeamA = []
Ft_TeamA = []
Oreb_TeamA = []
Dreb_TeamA = []
Reb_TeamA = []
Ast_TeamA = []
Pf_TeamA = []
To_TeamA = []
Stl_TeamA = []
Blk_TeamA = []

Name_TeamB =[]
Min_TeamB = []
Pts_TeamB = []
Fg_TeamB = []
F2g_TeamB = []
F3g_TeamB = []
Ft_Teamb = []
Oreb_TeamB = []
Dreb_TeamB = []
Reb_TeamB = []
Ast_TeamB = []
Pf_TeamB = []
To_TeamB = []
Stl_TeamB = []
Blk_TeamB = []
```

```python
for i in range(0,len(url_list)):
    print(url_list[i])
    driver.get(url_list[i])
    time.sleep(3)
    r_stats = driver.page_source
    soup = bs4.BeautifulSoup(r_stats, "lxml")
    
    timing = soup.select("#gamepage_header_game_post > div > div.header-scores_desktop > div.game-info > div.phase-info > span.phase")[0].text
    Timing.append(timing)
    
    name_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > header > span")[0].text
    Name_TeamA.append(name_teamA)
    
    min_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.min")[0].text
    Min_TeamA.append(min_teamA)
    
    pts_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.pts")[0].text
    Pts_TeamA.append(pts_teamA)
    
    fg_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.field-goals > span.made-all")[0].text
    Fg_TeamA.append(fg_teamA)
    
    f2g_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.field-goals-2p > span.made-all")[0].text
    F2g_TeamA.append(f2g_teamA)
    
    f3g_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.field-goals-3p > span.made-all")[0].text
    F3g_TeamA.append(f3g_teamA)
    
    ft_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.free-throw > span.made-all")[0].text
    Ft_TeamA.append(ft_teamA)
    
    oreb_teamA =  soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.reb-offence")[0].text
    Oreb_TeamA.append(oreb_teamA)
    
    dreb_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.reb-defence")[0].text
    Dreb_TeamA.append(dreb_teamA)
    
    reb_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.reb-total")[0].text
    Reb_TeamA.append(reb_teamA)
    
    ast_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.assists")[0].text
    Ast_TeamA.append(ast_teamA)
    
    pf_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.personal-fouls")[0].text
    Pf_TeamA.append(pf_teamA)
    
    to_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.turnovers")[0].text
    To_TeamA.append(to_teamA)
    
    stl_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.steals")[0].text
    Stl_TeamA.append(stl_teamA)
    
    blk_teamA = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-A > div > table > tfoot > tr.team-totals > td.block-shots")[0].text
    Blk_TeamA.append(blk_teamA)
    
    name_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > header > span")[0].text
    Name_TeamB.append(name_teamB)
    
    min_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.min")[0].text
    Min_TeamB.append(min_teamB)
    
    pts_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.pts")[0].text
    Pts_TeamB.append(pts_teamB)
    
    fg_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.field-goals > span.made-all")[0].text
    Fg_TeamB.append(fg_teamB)
    
    f2g_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.field-goals-2p > span.made-all")[0].text
    F2g_TeamB.append(f2g_teamB)
    
    f3g_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.field-goals-3p > span.made-all")[0].text
    F3g_TeamB.append(f3g_teamB)
    
    ft_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.free-throw > span.made-all")[0].text
    Ft_Teamb.append(ft_teamB)
    
    oreb_teamB =  soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.reb-offence")[0].text
    Oreb_TeamB.append(oreb_teamB)
    
    dreb_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.reb-defence")[0].text
    Dreb_TeamB.append(dreb_teamB)
    
    reb_teamb = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.reb-total")[0].text
    Reb_TeamB.append(reb_teamb)
    
    ast_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.assists")[0].text
    Ast_TeamB.append(ast_teamB)
    
    pf_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.personal-fouls")[0].text
    Pf_TeamB.append(pf_teamB)
    
    to_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.turnovers")[0].text
    To_TeamB.append(to_teamB)
    
    stl_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.steals")[0].text
    Stl_TeamB.append(stl_teamB)
    
    blk_teamB = soup.select("#gamepage_boxscore > div.tab_ajax_content > div > section.box-score_team-B > div > table > tfoot > tr.team-totals > td.block-shots")[0].text
    Blk_TeamB.append(blk_teamB)
    
    
```

```python
df_boxscore = pd.DataFrame({
    "Timing":Timing,
    "Name_TeamA":Name_TeamA,
    "Min_TeamA":Min_TeamA,
    "Pts_TeamA":Pts_TeamA,
    "Fg_TeamA":Fg_TeamA,
    "F2g_TeamA":F2g_TeamA,
    "F3g_TeamA":F3g_TeamA,
    "Ft_TeamA":Ft_TeamA,
    "Oreb_TeamA":Oreb_TeamA,
    "Dreb_TeamA":Dreb_TeamA,
    "Reb_TeamA":Reb_TeamA,
    "Ast_TeamA":Ast_TeamA,
    "Pf_TeamA":Pf_TeamA,
    "To_TeamA":To_TeamA,
    "Stl_TeamA":Stl_TeamA,
    "Blk_TeamA":Blk_TeamA,
    "Name_TeamB": Name_TeamB,
    "Min_TeamB":Min_TeamB,
    "Pts_TeamB":Pts_TeamB,
    "Fg_TeamB":Fg_TeamB,
    "F2g_TeamB":F2g_TeamB,
    "F3g_TeamB":F3g_TeamB,
    "Ft_TeamB":Ft_Teamb,
    "Oreb_TeamB":Oreb_TeamB,
    "Dreb_TeamB":Dreb_TeamB,
    "Reb_TeamB":Reb_TeamB,
    "Ast_TeamB":Ast_TeamB,
    "Pf_TeamB":Pf_TeamB,
    "To_TeamB":To_TeamB,
    "Stl_TeamB":Stl_TeamB,
    "Blk_TeamB":Blk_TeamB
})
```

```python
df_boxscore.to_csv("worldcup_boxscore.csv")
```
